다차원 배열을 tensor라고도 한다.  
numpy의 ndarray 인스턴스에는 ndim이라는 인스턴스 변수가 있다. 
- ndim: 'number of dimensions'의 약자. 다차원 배열의 '차원 수'

```python
import numpy as np
x = np.array(1)
x.ndim
```

```python
x = np.array([1,2,3])
x.ndim
```

```python
x = np.array([[1, 2, 3],
              [4, 5, 6]])
x.ndim
```
